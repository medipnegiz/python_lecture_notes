```python
liste = [9, 2, 8, 4, 7]
harf = ['A', 'y', 'Z', 'a', 'c']
liste.insert(2,5)                   #   istedigimiz yere ekleme yapar ve diger elemani bir saga kaydirir.
liste.insert(len(liste), 6)         #   sona ekleme yapmak icin len fonksiyonu kullanilabilir 
liste.remove(4)                     #   liste icinden belirtilen elemani siler
liste.sort()                        #   kücükten büyüge dogru siralama yapar. "ayni tip olmasi lazim"
liste.sort(reverse=True)            #   önce kücükten büyüge dogru siralar sonra tersten yazdirma islemi yapar
harf.sort()                         #   ASCII kod tablosuna göre siralar
print(liste)
print(harf)
```

```python
# listeden matris tanimlamasi yaptik
matrix = [
          [1,2,3,4],
          [5,6,7,8],
          [9,0,1,2],
         ]
b = list(zip(* matrix))                             #   satirlari ve sütunlari birlestirerek yazar
print(b)
lnew = matrix[0] + matrix[1] + matrix[2]            #   listeleri birlestirdik
t, k ,l = matrix                                    #   listeleri birlestirmenin bir baska yolu
print(t+k+l)                                        
print(* matrix)
print(lnew)
```

```python
dizi = ['a', 'b', 'c', 'd', 'e']
dizi.reverse()
ndizi = ''.join(dizi)       #   aradaki karakterleri kaldirdi tamamen birlestirdi
print(*dizi)                #   köseli parantezden ve virgülden kurtarmak icin * kullanilir
dizi.pop(4)                 #   verilen sayidaki degeri bulur ve siler (sayi verilmezse en son karakteri siler)
#dizi.remove('c')           #   istenilen karakteri silme
del dizi[0]                 #   silme islemi icin kullandik
#dizi.clear()               #   listenin icini tamamen temizlemek icin kullanilir
print(dizi)
print(ndizi)
```

```python
# Tuple degistirilemezler ve Listelerden daha hizli calisirlar
t = ()
t1 = tuple()
t2 = 'p', 'y', 't', 'h', 'o', 'n'           #   virgüller ile ayrildigi icin "tuple" ifadedir
t3 = ('p', 'y', 't', 'h', 'o', 'n')
t4 = 'Python'
t5 = 1, 2, 3
t6 = ('a')                                  #   parantez icinde olsada virgül olmazsa "string" ifade olarak algilar
t7 = ('a',)                                 #   virgul oldugu icin "tuple" olarak algilar
print(type(t))
print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))
print(type(t6))
print(type(t7))
```
