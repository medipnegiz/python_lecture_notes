```python
# all (iterable)yenilenebilir degerlerde yada bos degerlerde True döndürür, digerlerinde False döndürür. (üzerinde sirayla gidebilecegimiz seyler iterabledir)
# tüm elemanlar True sonuc üretirse True degerini verir. "And" mantigi ile calisir
names = ['Susan', 'Tom', 'False']
nn = ['Susan', 'Tom', False]
mood = ['happy', 'saf', 0]
mm = ['müde', 5] # 5: True
empty = {}
#bos = 46513    #  iterable olmadigi icin hata verir
bos = '46798'
print(all(names), all(nn), all(mood), all(empty), all(mm), all(bos), sep='\n')
```

```python
# herhangi bir deger True ise deger True döndürür. "Or" mantigi ile calisir
names = ['Susan', 'Tom', False]
nn = [None, (), False]
mm = ['müde', (3,)] # 5: True
empty = {}
print(any(names), any(nn), any(empty), any(mm), sep='\n')
```

```python
n = int(input('Bir sayi girin: '))
def prime(n):
    flag = True
    if n < 2:   # asal
        flag = True
    for i in range(2,int(pow(n, 0.5))+1):
        if n % i == 0:  #   asal degil
            flag = False
    return flag
prime(n)
print(list(filter(prime, range(2,n))))
```
