```python
def calculator(a, b, c):
    if c == '+':
        return (a + b)
    elif c == '-':
        return (a - b)
    elif c == '*':
        return (a * b)    
    elif c == '/':
        return (a / b)
    else:
        return ('Gecerli bir argüman giriniz...')
sonuc = calculator(9, 4, '+')
print(sonuc)
```

```python
## mutlak deger döndürme, negatif sayilari pozitif cevirme
def my_abs(n):
    '''
    Bu fonksiyon mutlak deger alir
    '''
    if n < 0:
        return -n       #   sayi negatif ise eksi ile carparak pozitif yaptik
    else:
        return n
print(my_abs(-5))
print(my_abs(7))
print(my_abs.__doc__)                   #   "__doc__" fonksiyon icindeki string ifadeleri yazdirmak icin kullanilir
```

```python
## eger fonksiyon kendini cagiriyorsa veya kendine atifta bulunuyorsa bu fonksiyon recursive(özyinelemeli) fonksiyondur

def fakt(n):
    if n <= 1:
        return 1
    else:
        return n * fakt(n-1)            #   burada fonksiyon kendisini cagirdi ve islem yapti
print(fakt(5))
```

```python
## döngüsel yaklasim ile fibonacci sayisini buldurma

def fib(n):
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b       #   yer degistirme islemi yapiyor
    print()
fib(100)
```

```python
## fibonacci serisinin recorcive cözümü **önemli olan duracagi yeri tanimlamak**

def fib(n):
    if n <= 1:      ## -->  if n in {0,1} ***
        return n
    else:
        return fib(n-1) + fib(n-2)
for i in (range(15)):                   #   burada 15 tane fibonacci sayisi göstermesini sagladik
    print(fib(i), end=' ')
```

```python
## recorcive fonksiyonu ile üs alma fonksiyonu 2^n

def us(n):
    if n == 0:
        return 1
    else:
        return 2 * us(n-1)      #   burada program sürekli kendini cagirarak islem yapiyor
print(us(5))
```

```python
def fruit(a, b):
    print(f'I want {a} and {b}')
fruit('portakal', 'karpuz')

def fruit(*a):
    print(f'I want {a[0]} and {a[1]}')
fruit('portakal', 'karpuz')

def meyve(*c):
    t = 'I want '
    for i in c:
        t = t + i + ' and '
    print(t.rstrip(' and '))
meyve('Elma', 'Cilek', 'Erik')
```

```python
def slicer(*a):         #   *args degisken olarak tanimlandi
    tek = []
    cift = []
    for i in a:
        if i % 2 == 0:
            cift.append(i)
        else:
            tek.append(i)
    print('Tek sayilar: ', tek)
    print('Cift sayilar: ', cift)
slicer(1,2,3,4,5,6,7,8,9)
```

```python
def tek_cift(*x):                           #   *args degisken olarak tanimlandi
    tek = [i for i in x if i % 2 == 1]
    cift = [i for i in x if i % 2 == 0]
    print('Tek sayilar: ', tek, '\nCift sayilar: ', cift)
tek_cift(1,2,3,4,5,6,7,8,9)
```

```python
def liste(**kwargs):
    name = []
    yas = []
    for i, j in kwargs.items():
        name.append(i)
        yas.append(j)
    print(name, yas, sep='\n')
liste(ali=33, ufuk=31, murat= 29) 
```

```python
def liste(**kwargs):
    name = [i for i in kwargs.keys()]
    yas = [k for k in kwargs.values()]
    print(name, yas, sep='\n')
liste(ali=33, ufuk=31, murat= 29)
```
