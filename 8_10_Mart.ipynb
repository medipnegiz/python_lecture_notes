```python
import random
print('... Aklimda tuttugum sayiyi tahmin et ...')
a = random.randint(1,10)
sayac = 0
puan = 100
while True:
    sayi = int(input('sayiyi girin..: '))
    sayac += 1
    puan -= 10
    if puan < 70:
        print('kaybettin...')
        break
    elif sayi < a:
        print('Daha büyük ')
    elif sayi > a:
        print('Daha kücük ')
    else:
        print(f'Tahmin sayi: {a}')
        print(f'{sayac} tahminde bildin')
        break
print(f'Toplam skor: {puan}')
```

```python
cumle = 'Ali uzunkavakaltindan uyumazoglu'
word = cumle.split()                        #   cümleyi kelimelere böldü ve liste icine aldi
print(word)
s = 0
kelime = 0
while s < len(word):
    if len(word[s]) > kelime:
        kelime = len(word[s])               #   en uzun kelime icin atama yapti
    s += 1
print(f'en uzun kelimenin uzunlugu {kelime}')
```

```python
# pythonic bir cozüm olarak daha hizli calisir
cumle = 'Ali uzunkavak altindan uyumazoglu'
word = cumle.split()                       
print(word)
kelime_uzun = len(max(word, key = len))             #   max komutu icine yeni bir "key" alarak onun en büyük olanini verebilir
kelime = max(word, key=len)
print(f'en uzun kelime {kelime_uzun} karakter')
print(f'en uzun kelime {kelime}')
```

### **RANDOM KOMUTU**

```python
import random
a = random.randint(1,20)                                            #   rastgele bir tamsayi üretmek icin kullanilir
b = random.uniform(0.5, 1.5)                                        #   rastgele bir kesirli sayi üretecek
c = random.random()                                                 #   rastgele bir sayi üretir

liste = ['ali', 'veli', 'ahmet', 'mehmet', 'celal']
d = random.choice(liste)                                            #   liste arasindan rastgele bir ifade secer

kardiz = 'istihza'
e = random.choice(kardiz)                                           #   kelime arasindan rastgele secim yapar

l = list(range(10))
random.shuffle(l)                                                   #   listeyi karisik bir sekilde yazdirir

lis = range(100)
k= random.sample(lis, 5)                                            #   belirli degerler arasinda istedigimiz kadar rastgele sayi yazdirir

print(a)
print(b)
print(c)
print(d)
print(e)
print(l)
print(k)
```
