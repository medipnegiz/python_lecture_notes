```python
import random
print('Sayisal Loto')
loto = []
for i in range(6):
    sayi = random.randint(1,49)
    if sayi in loto:
        sayi = random.randint(1,49)
    else:
        loto.append(sayi)
print(loto)
print()
```

```python
# sifre üretme programi, belirli sayida sayi, harf ve özel karakter icerir
import string 
from random import sample

k = string.ascii_lowercase
b = string.ascii_uppercase
r = string.digits
o = string.punctuation
p = string.printable
key = sample(k,1) + sample(b,1) + sample(r,1) + sample(o,1) + sample(p, 6)
random.shuffle(key)
print(''.join(key))
```

```python
# dogum gününe kac gün kalindigini bulduk
import datetime
b = datetime.date.today()
d = datetime.date(b.year, 4, 13)
print(b)
if d < b:
    d = d.replace(year=b.year+1)
print(f'dogum gününe {abs(d-b).days} gün kaldi')
```

```python
from datetime import date
dogum = date(1974, 6, 15)
olum = date(2042, 11, 21)
life = olum - dogum
print(f'{life.days} gün yasamistir')
```

```python
from datetime import date
dogum = date(1974, 6, 15)
olum = date(2042, 11, 21)
life = date.toordinal(olum) - date.toordinal(dogum)
print(f'{life} gün yasamistir ')
```
