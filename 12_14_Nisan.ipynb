```python
while True:
    try:
        sayi1 = int(input('Ilk sayiyi girin..:'))
        sayi2 = int(input('Ikinci sayiyi girin..:'))
        bol = sayi1 / sayi2
        print('Bölüm sonucu..:', bol)
        break
    #except Exception:                                   #   bütün exception hatalari icin ortak bir mesaj verilebilir
    except ZeroDivisionError:                            #   sayi 0'a bölünmez hatalari icin kullanilir
        print('Bölen 0 olamaz...')
    except TypeError:                                    #   type hatalari icin kullanilir
        print('Yanlis giris, lütfen rakam girin...')
    except ValueError:                                   #   Value hatalari icin kullanilir
        print('Yanlis giris, lütfen rakam girin...')
```

```python
try:
    dd = open('my_file.txt', 'r')
    print(dd.read())
except FileNotFoundError:                   #   dosya bulunamadi hatasi
    print('Böyle bir dosya yok')
finally:                                    #   bu kisim her türlü calisacaklari icerir
    #dosya.close()
    print('Dosya kapatildi')
```

```python
fruits = ['banana', 'mango', 'pear', 'apple', 'kiwi', 'grape']
while True:
    try:
        gir = int(input('Indis numarasi giriniz..:'))
    except (ValueError, TypeError):
        print('lütfen rakam girin...')
    except IndexError:
        print('0 - 5 arasinda rakam girin...')
    else:
        print(f'My favorite is {fruits[gir]}')
        break
    finally:
        print('Meyvelerimiz hep tazedir...')
```

```python
fruits = ['banana', 'mango', 'pear', 'apple', 'kiwi', 'grape']
n = 3
while n > 0:
    try:
        gir = int(input('Indis numarasi giriniz..:'))
        print(f'My favorite is {fruits[gir]}')
    except (ValueError, TypeError):
        n -= 1
        print(f'lütfen rakam girin!')
        if n == 0:
            print('Hakkiniz kalmadi!!!')
        else:
            print(f'{n} hakkiniz kaldi!')
    except IndexError:
        n -= 1
        print('0 - 5 arasinda rakam girin!')
        if n == 0:
            print('Hakkiniz kalmadi!!!')
        else:
            print(f'{n} hakkiniz kaldi!')
    else:
        print(f'Tebrikler dogru deger girildi')
        break
```

```python
d = open('../myfile.txt', 'r', encoding='utf-8')        #   dosyanin yolu bu sekilde de yazilabilir
print(d.read())
print(d.read(5))                                        #   dosya icerisindeki ilk 5 karakteri yazdirir
print(d.readline())                                     #   satir satir okur ve her yazdirmada bir satir yazdirir
print(d.readlines())                                    #   Tüm satirlari okur ve liste halinde yazdirir
d.close()
```

```python
d = open('../myfile.txt', 'r', encoding='utf-8')
for i in d:
    print(i)                            #   Tüm satirlari yazdirmak icin kullanilir
d.close()
```

```python
with open('fishes.txt', 'r', encoding='utf-8') as f:
    l = f.read(15)
    while len(l) > 0:
        print(l, end='*')
        l = f.read(15)
```

```python
## parca parca okuduklarimizi listeye ekledik
with open('fishes.txt', 'r', encoding='utf-8') as f:
    d = []
    while True:
        l = f.read(10)
        if l == '':
            break
        else:
            d.append(l)
print(d)
```

```python
## lambda fonksiyonu ile parca parca ekleme yaptik
with open('fishes.txt', 'r', encoding='utf-8') as f:
    d = [l for l in iter(lambda: f.read(10), '')]
print(d)
```

YAZMA MODU

```python
with open('test1.txt', 'w', encoding='utf-8') as f:             #   "w" ile yazma moduna aldik
    t = 'Ne olursan ol, yinede gel...'
    f.write(t)                                                  #   yazma islemi burada gerceklesiyor

with open('test1.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
fruits = ['Banana', 'Orange', 'Apple', 'Strawberry', 'Cherry']
with open('test1.txt', 'w', encoding='utf-8') as f:  
    for i in fruits:
        f.write(i + '\n')

with open('test1.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
with open('test1.txt', 'r', encoding='utf-8') as f:                 #   bir dosyayi önce actik ve okuduk
    with open('test2.txt', 'w', encoding='utf-8') as d:             #   dosyanin icine yazmak icin tekrar yazma modunda actik
        d.write(f.read())                                           #   acilan ve okunan bir dosyayi baska dosya icine yazdirdik

with open('test2.txt', 'r', encoding='utf-8') as e:
    print(e.read())
```

```python
with open('car.jpg', 'rb') as f:                    #   resim dosyasini okumak icin "b" binary modu "r" ile kullanilir
    with open('c_car.jpg', 'wb') as d:              #   resim dosyasini yazmak icin "b" binary modu "w" ile kullanilir
        d.write(f.read())                           #   farkli bir isimde dosyayi yeniden olusturdu
```

```python
fruits = ['Banana\n', 'Orange\n', 'Apple\n', 'Strawberry\n', 'Cherry\n']
with open('test2.txt', 'w', encoding='utf-8') as f: 
    f.writelines(fruits)                                #   listeyi string formatinda ekliyor

with open('test2.txt', 'r', encoding='utf-8') as d: 
    print(d.readlines())
    d.seek(0)
    print(d.read())
```

```python
fruits = ['Banana\n', 'Orange\n', 'Apple\n', 'Strawberry\n', 'Cherry\n']
with open('test2.txt', 'a', encoding='utf-8') as f:     #   altina ekleme yapmak icin "a" formatinda yazdik
    f.writelines('Melone')                              #   istenilen string ifadeyi yazdirma yaptik

with open('test2.txt', 'r', encoding='utf-8') as d: 
    print(d.read())
```

```python
sozluk = {1:'birinci', 2:'ikinci', 3: 'ücüncü', 4:'dördüncü', 5:'besinci'}
with open('test3.txt', 'w') as f:
    for i in sozluk.keys():
        f.write('Bu ' + sozluk[i] + ' satirdir.\n')
with open('test3.txt', 'r') as f:
    print(f.read())
    print(f.mode)
```
